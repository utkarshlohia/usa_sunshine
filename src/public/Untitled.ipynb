```python
file_path = '/Users/utkarshlohia/Desktop/archive (1)/daily_weather.parquet'

# Read the Parquet file into a DataFrame
raw_df = pd.read_parquet(file_path, engine='pyarrow')  # or engine='fastparquet'
```

```python
raw_df['date'] = pd.to_datetime(raw_df['date'])
```

```python
cities_df = pd.read_csv('/Users/utkarshlohia/Desktop/archive (1)/cities.csv')
cities_df
```

```python
new_raw_df = raw_df[(raw_df['date'].dt.year > 2000)]
new_raw_df = new_raw_df[(new_raw_df['date'].dt.year < 2023)]
merged_df = new_raw_df.merge(cities_df, on='city_name')
merged_df = merged_df[merged_df['country'] == 'India']
```

```python
new_merged_df = merged_df[['city_name', 'date', 'sunshine_total_min', 'latitude', 'longitude']]
```
